Python hỗ trợ nhiều kiểu dữ liệu cơ bản, bao gồm:

Kiểu số (Numeric Types):
+ int: Số nguyên (ví dụ: 10, -5, 0)
+ float: Số thực (ví dụ: 3.14, -0.5, 2.0)
+ complex: Số phức (ví dụ: 1 + 2j)

Kiểu chuỗi (Sequence Types):
+str: Chuỗi ký tự (ví dụ: "hello", 'Python')
+list: Danh sách có thứ tự, có thể thay đổi (ví dụ: [1, 2, 3], ['apple', 'banana'])
+tuple: Bộ giá trị có thứ tự, không thể thay đổi (ví dụ: (10, 20, 30), ('a', 'b'))
+range: Dãy số không đổi (thường dùng trong vòng lặp)

Kiểu ánh xạ (Mapping Type):
+dict: Từ điển, lưu trữ dữ liệu dưới dạng cặp khóa-giá trị (ví dụ: {'name': 'Alice', 'age': 30})
Kiểu tập hợp (Set Types):
+set: Tập hợp không có thứ tự, không chứa các phần tử trùng lặp (ví dụ: {1, 2, 3})
+frozenset: Giống như set nhưng không thể thay đổi

Kiểu Boolean:
+bool: Giá trị logic, chỉ có hai giá trị là True hoặc False

Kiểu None:
+NoneType: Biểu thị sự vắng mặt của giá trị (chỉ có một giá trị là None)

Câu 5:
Trong Python, có hai loại ghi chú chính:

1. Ghi chú dòng đơn (Single-line comments):
Bắt đầu bằng ký tự #.
Mọi thứ sau # trên cùng một dòng sẽ được trình thông dịch Python bỏ qua.
Thường được sử dụng để giải thích một dòng mã cụ thể hoặc thêm ghi chú ngắn.
2. Ghi chú nhiều dòng (Multi-line comments) hoặc Docstrings:
Mặc dù Python không có cú pháp ghi chú nhiều dòng riêng biệt như một số ngôn ngữ khác, người ta thường sử dụng chuỗi ký tự được đặt trong ba dấu nháy đơn ('''...''') hoặc ba dấu nháy kép ("""...""") để tạo ghi chú nhiều dòng.
Nếu những chuỗi ký tự này xuất hiện ngay sau định nghĩa của một hàm, lớp, mô-đun hoặc phương thức, chúng được gọi là "docstrings". Docstrings được sử dụng để tài liệu hóa mã nguồn và có thể được truy cập bằng thuộc tính __doc__ hoặc hàm help().
Nếu chúng không ở vị trí docstring, chúng đơn giản là các chuỗi ký tự không được gán cho biến nào, và trình thông dịch Python sẽ bỏ qua chúng.



Câu 6:
_(Toán tử chia):
Thực hiện phép chia và trả về kết quả là một số thực (float), kể cả khi kết quả là số nguyên.
Ví dụ: 10 / 2 trả về 5.0, 7 / 3 trả về 2.333...

_(Toán tử chia lấy phần nguyên):
Thực hiện phép chia và trả về phần nguyên của kết quả. Phần thập phân bị loại bỏ.
Ví dụ: 10 // 3 trả về 3, -10 // 3 trả về -4 (làm tròn xuống số nguyên gần nhất nhỏ hơn hoặc bằng kết quả).

_(Toán tử chia lấy phần dư - Modulo):
Thực hiện phép chia và trả về phần dư của phép chia đó.
Ví dụ: 10 % 3 trả về 1, 15 % 4 trả về 3.
** (Toán tử lũy thừa):
Thực hiện phép nâng một số lên lũy thừa của một số khác.
Ví dụ: 2 ** 3 trả về 8 (2 mũ 3), 9 ** 0.5 trả về 3.0 (căn bậc hai của 9).
and (Toán tử logic AND):
Kết hợp hai biểu thức boolean. Trả về True nếu cả hai biểu thức đều là True. Trả về False nếu ít nhất một trong hai biểu thức là False.
Ví dụ: (5 > 3) and (10 < 20) trả về True, (5 > 3) and (10 > 20) trả về False.

_(Toán tử logic OR):
Kết hợp hai biểu thức boolean. Trả về True nếu ít nhất một trong hai biểu thức là True. Chỉ trả về False khi cả hai biểu thức đều là False.
Ví dụ: (5 > 3) or (10 > 20) trả về True, (5 < 3) or (10 > 20) trả về False.

_(Toán tử nhận dạng):
Kiểm tra xem hai biến có cùng tham chiếu đến cùng một đối tượng trong bộ nhớ hay không. Khác với toán tử == (so sánh giá trị).

Câu 8:
1. Lỗi cú pháp (Syntax Errors): Đây là những lỗi xảy ra khi mã nguồn của bạn không tuân thủ các quy tắc ngữ pháp của ngôn ngữ Python. Trình thông dịch Python sẽ phát hiện ra lỗi này ngay trước khi chương trình được thực thi.

2. Lỗi ngữ nghĩa (Semantic Errors) hoặc Lỗi logic (Logic Errors): Loại lỗi này xảy ra khi cú pháp đúng, chương trình chạy được, nhưng kết quả không như mong đợi. Đây là lỗi khó phát hiện nhất vì trình thông dịch không báo lỗi.

Ví dụ: Sử dụng sai công thức tính toán, điều kiện trong if sai, lặp sai số lần.
Cách phát hiện: Chạy thử chương trình với nhiều bộ dữ liệu khác nhau và kiểm tra kết quả. Sử dụng các công cụ gỡ lỗi (debugger) để theo dõi luồng thực thi và giá trị biến.

3. Lỗi ngoại lệ (Exceptions) hoặc Lỗi runtime (Runtime Errors): Đây là những lỗi xảy ra trong quá trình chương trình đang chạy, khi Python gặp phải một tình huống mà nó không thể xử lý được. Khác với lỗi cú pháp, lỗi ngoại lệ chỉ xuất hiện khi chương trình thực sự chạy đến đoạn mã gây lỗi.

Ví dụ: Chia cho 0 (ZeroDivisionError), truy cập vào phần tử không tồn tại trong danh sách (IndexError), cố gắng chuyển đổi một chuỗi không phải số sang số (ValueError), mở một tập tin không tồn tại (FileNotFoundError).
Nếu không được xử lý, lỗi ngoại lệ sẽ làm chương trình dừng đột ngột và in ra thông báo lỗi (traceback).

```python
# Các biến ví dụ (bạn có thể thay đổi giá trị)
x = 10
y = 5
ncc = 3
number_of_closed_cases = 20

print("--- Lệnh ban đầu ---")
# (a) x = x + 1
x_a_original = x
x_a_original = x_a_original + 1
print(f"(a) x = x + 1 -> Ket qua x = {x_a_original}")

# (b) x = x / 2
x_b_original = x
x_b_original = x_b_original / 2
print(f"(b) x = x / 2 -> Ket qua x = {x_b_original}")

# (c) x = x – 1
x_c_original = x
x_c_original = x_c_original - 1
print(f"(c) x = x – 1 -> Ket qua x = {x_c_original}")

# (d) x = x + y
x_d_original = x
x_d_original = x_d_original + y
print(f"(d) x = x + y -> Ket qua x = {x_d_original}")

# (e) x = x - (y + 7)
x_e_original = x
x_e_original = x_e_original - (y + 7)
print(f"(e) x = x - (y + 7) -> Ket qua x = {x_e_original}")

# (f) x = 2*x
x_f_original = x
x_f_original = 2 * x_f_original
print(f"(f) x = 2*x -> Ket qua x = {x_f_original}")

# (g) number_of_closed_cases = number_of_closed_cases + 2*ncc
number_of_closed_cases_g_original = number_of_closed_cases
number_of_closed_cases_g_original = number_of_closed_cases_g_original + 2 * ncc
print(f"(g) number_of_closed_cases = number_of_closed_cases + 2*ncc -> Ket qua number_of_closed_cases = {number_of_closed_cases_g_original}")


print("\n--- Lệnh viết ngắn gọn ---")
# (a) x += 1
x_a_short = x
x_a_short += 1
print(f"(a) x += 1 -> Ket qua x = {x_a_short}")

# (b) x /= 2
x_b_short = x
x_b_short /= 2
print(f"(b) x /= 2 -> Ket qua x = {x_b_short}")

# (c) x -= 1
x_c_short = x
x_c_short -= 1
print(f"(c) x -= 1 -> Ket qua x = {x_c_short}")

# (d) x += y
x_d_short = x
x_d_short += y
print(f"(d) x += y -> Ket qua x = {x_d_short}")

# (e) x -= (y + 7)
x_e_short = x
x_e_short -= (y + 7)
print(f"(e) x -= (y + 7) -> Ket qua x = {x_e_short}")

# (f) x *= 2
x_f_short = x
x_f_short *= 2
print(f"(f) x *= 2 -> Ket qua x = {x_f_short}")

# (g) number_of_closed_cases += 2*ncc
number_of_closed_cases_g_short = number_of_closed_cases
number_of_closed_cases_g_short += 2 * ncc
print(f"(g) number_of_closed_cases += 2*ncc -> Ket qua number_of_closed_cases = {number_of_closed_cases_g_short}")
```
